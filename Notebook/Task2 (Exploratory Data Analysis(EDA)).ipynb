## Level 1: Task 2(Exploratory Data Analysis(EDA))

```python
#load Dataset & inspect
import pandas as pd

# Loading the cleaned CSV
df = pd.read_csv("cleaned_house_prediction_data.csv")
df.head()  
```

#### Summary Statistics

```python
summary_stats = df.describe().T
summary_stats['median'] = df.median()
summary_stats['mode'] = df.mode().iloc[0]

summary_stats
```

#### Data Distribution (Histograms)

```python
import matplotlib.pyplot as plt

df.hist(figsize=(15,10), bins=20, edgecolor='black')
plt.tight_layout()
plt.show()
```

#### Boxplots (Outliers)

```python
import seaborn as sns

features = ['rooms', 'low_income', 'crime_rate', 'price']

for feature in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

#### Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

#### Scatterplots

```python
#Scatterplots for Important Features
import matplotlib.pyplot as plt
import seaborn as sns

# List of top features affecting Price
top_features = ['rooms', 'low_income', 'student_teacher_ratio', 'crime_rate']

# Plot scatterplots
for feature in top_features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'Price vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```

### Insights from EDA

- Houses with more rooms tend to have higher prices (positive relationship).
- Areas with higher low income population have lower house prices (negative relationship).
- Higher crime rate is associated with lower house prices.
- Student-teacher ratio also shows a slight negative impact on price.
- The correlation matrix confirms strong relationships between some features and price.
- Boxplots show presence of outliers in features like price and low_income.
- Histograms indicate that some features are not normally distributed (skewed data).
